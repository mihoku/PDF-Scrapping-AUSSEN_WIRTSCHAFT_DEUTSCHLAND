```python
def read_pdf(filename, start, end):
    pdfFileObj = open(filename, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    extracted_text=[]
    for i in range(start,end):
        pageObj = pdfReader.getPage(i)
        extracted_text.append(pageObj.extractText())
    pdfFileObj.close()
    return extracted_text
```

```python
def text_splitter(case_text, keyword_start,keyword_limit):
    text_result = case_text[case_text.find(keyword_start)+len(keyword_start):case_text.find(keyword_limit)]
    return text_result
```

```python
def get_nutzung(case_text):
    return text_splitter(case_text,'NUTZUNG','KENNDATEN')

def get_name(case_text):
    text_list = case_text[:case_text.find('NUTZUNG')].split('\n')
    return text_list[len(text_list)-2]

def get_page(case_text):
    return case_text.split('\n')[0]

def get_kenndaten(case_text):
    text_list = text_splitter(case_text,'KENNDATEN','LAGE')
    text = ''
    for i in text_list:
        text+=i
    return text

def get_lage(case_text):
    return text_splitter(case_text,'LAGE','VORAUSSICHTLICHE FERTIGSTELLUNG')

def get_website(case_text):
    text_list = text_splitter(case_text,'WEITERE INFORMATIONEN FINDEN SIE HIER','(Quelle zuletzt').split('\n')
    return text_list

def has_numbers(inputString):
    return bool(re.search(r'\d', inputString))

def get_vf(case_text):
    text_picked = text_splitter(case_text,'VORAUSSICHTLICHE FERTIGSTELLUNG','WEITERE INFORMATIONEN FINDEN SIE HIER')
    uppercase_list = []
    for x in text_picked.split('\n'):
        if x.isupper() and bool(re.search(r'\d', x))==False:
            uppercase_list.append(x)
    uppercase_list = [item for item in uppercase_list if 'OFFEN' not in item]
    upper_limit = uppercase_list[0]
    text_list = text_picked[:text_picked.find(upper_limit)].split('\n')
    text = ''
    for i in text_list:
        text+=i
    return text

def get_contact_section(case_text):
    text_picked = text_splitter(case_text,'VORAUSSICHTLICHE FERTIGSTELLUNG','WEITERE INFORMATIONEN FINDEN SIE HIER')
    uppercase_list = []
    for x in text_picked.split('\n'):
        if x.isupper()  and bool(re.search(r'\d', x))==False:
            uppercase_list.append(x)
    uppercase_list = [item for item in uppercase_list if 'OFFEN' not in item]
    limit = uppercase_list[0]
    text_picked = text_picked[text_picked.find(limit):]
    return text_picked

def extract_contact(case_text):
    text_list = case_text.split('\n')
    contact_list=[]
    txt = ''
    for i in range(len(text_list)):
        if text_list[i][:2]=='W ':
            if i<len(text_list)-1:
                if text_list[i+1].isupper():
                    txt = txt+text_list[i]+'\n'
                    contact_list.append(txt)
                    txt = ''
                else:
                    txt = txt+text_list[i]+'\n'
            else:
                txt = txt+text_list[i]+'\n'
                contact_list.append(txt)
                txt = ''
        else:
            txt = txt+text_list[i]+'\n'
    return contact_list
    

def get_contact(text_picked):
    text_list = text_picked.split('\n')
    separator = [0]
    for i in range(len(text_list)):
        if text_list[i].isupper() and '+' not in text_list[i] and text_list[i][:2]!='T ' and i!=separator[len(separator)-1]+1 and i!=separator[len(separator)-1]:
            separator.append(i)
    contact_list = []
    for i in range(len(separator)):
        if(i!=len(separator)-1):
            txt = ''
            for x in text_list[separator[i]:separator[i+1]]:
                txt=txt+x+'\n'
            contact_list.append(txt)
        else:
            txt = ''
            for x in text_list[separator[i]:]:
                txt=txt+x+'\n'
            contact_list.append(txt)
    return contact_list

def contact_list_generator(df):
    contact_list = []
    position_list = []
    name_list = []
    company_list = []
    address_list = []
    website_list = []
    phone_list = []
    email_list = []
    for x in df.index:
        contact_separated = get_contact(df['CONTACT_SECTION'][x])
        for i in contact_separated:
            contact_text = i.split('\n')
            position_list.append(contact_text[0])
            name_list.append(df['NAME'][x])
            company_list.append(contact_text[1])
            address_list.append(contact_text[2])
            txt = ''
            phone_txt = []
            website_txt = []
            email_txt = []
            for z in contact_text[2:]:
                txt+=z
                if z[:2]=='E ':
                    email_txt.append(z[2:])
                elif z[:2]=='T ':
                    phone_txt.append(z[2:])
                elif z[:2]=='W ':
                    website_txt.append(z[2:])
                if len(z[z.find('T '):z.find('E ')])!=0:
                    phone_txt.append(z[z.find('T ')+2:z.find('E ')])
                if len(z[z.find('E '):z.find('W ')])!=0:
                    email_txt.append(z[z.find('E ')+2:z.find('W ')])
            contact_list.append(txt)
            website_list.append(website_txt)
            phone_list.append(phone_txt)
            email_list.append(email_txt)
    dataset = pd.DataFrame()
    dataset['PROJECT'] = name_list
    dataset['ART'] = position_list
    dataset['NAME'] = company_list
    dataset['ADDRESS'] = address_list
    dataset['PHONE'] = phone_list
    dataset['EMAIL'] = email_list
    dataset['WEBSITE'] = website_list
    dataset['CONTACT_DETAILS'] = contact_list
    return dataset
        
    
```

```python
dataset['NUTZUNG'] = dataset.extracted_text.apply(get_nutzung)
dataset['PAGE'] = dataset.extracted_text.apply(get_page)
dataset['NAME'] = dataset.extracted_text.apply(get_name)
dataset['KENNDATEN'] = dataset.extracted_text.apply(get_kenndaten)
dataset['LAGE'] = dataset.extracted_text.apply(get_lage)
dataset['WEBSITE'] = dataset.extracted_text.apply(get_website)
dataset['VORAUSSICHTLICHE FERTIGSTELLUNG'] = dataset.extracted_text.apply(get_vf)
dataset['CONTACT_SECTION'] = dataset.extracted_text.apply(get_contact_section)
dataset
```
